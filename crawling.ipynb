```python
import time

# 최종 데이터를 저장할 리스트 초기화
final_data = []

flag = False
flag1 = False
flag2 = False

# 구 코드: 3023000000
# 구 코드: 3011000000
# 구 코드: 3017000000
# 구 코드: 3020000000
# 구 코드: 3014000000

# 대전광역시의 구 정보 가져오기
gungu_list = get_gungu_info(get_sido_info()[4])

for gungu in gungu_list:
    # 각 구에 속한 동 정보 조회
    print(f"구 코드: {gungu}")
    if gungu == "3020000000":
        flag = True
        continue

    if flag:
        dong_list = get_dong_info(gungu)
        
        for dong in dong_list:
            print(f"동 코드: {dong}")
            # 각 동에 있는 아파트 목록 조회
            apt_list = get_apt_list(dong)
            
            for apt in apt_list['complexList']:
                complex_no = apt['complexNo']
                
                print(f"아파트 코드: {complex_no}")
                article_list = get_article_list(complex_no)
                for article in article_list['articleList']:
                    article_no = article['articleNo']
                    
                    # 매물 상세 정보 조회
                    detail_info = get_detail(article_no, complex_no)
                    if 'articleDetail' in detail_info:
                        article_detail = detail_info['articleDetail']
                        
                        # 학군 정보 조회
                        school_info = get_school_info(complex_no)
                        
                        # 정보를 딕셔너리에 저장
                        try:
                            data_entry = {
                                "aptName": article_detail.get("aptName", "없음"),
                                "tradeBuildingTypeCode": article_detail.get("tradeBuildingTypeCode", "없음"),
                                "aptHeatMethodTypeName": article_detail.get("aptHeatMethodTypeName", "없음"),
                                "aptHeatFuelTypeName": article_detail.get("aptHeatFuelTypeName", "없음"),
                                "aptParkingCountPerHousehold": article_detail.get("aptParkingCountPerHousehold", "없음"),
                                "aptHouseholdCount": article_detail.get("aptHouseholdCount", "없음"),
                                "exposureAddress": article_detail.get("exposureAddress", "없음"),
                                "monthlyManagementCost": article_detail.get("monthlyManagementCost", "없음"),
                                "articleFeatureDescription": article_detail.get("articleFeatureDescription", "없음"),
                                "detailDescription": article_detail.get("detailDescription", "없음"),
                                "floorLayerName": article_detail.get("floorLayerName", "없음"),
                                "principalUse": article_detail.get("principalUse", "없음"),
                                "tagList": article_detail.get("tagList", []),
                                "schoolName": school_info.get("schools", [{}])[0].get("schoolName", "없음"),
                                "organizationType": school_info.get("schools", [{}])[0].get("organizationType", "없음"),
                                "establishmentYmd": school_info.get("schools", [{}])[0].get("establishmentYmd", "없음"),
                                "walkTime": school_info.get("schools", [{}])[0].get("walkTime", "없음"),
                                "studentCountPerTeacher": school_info.get("schools", [{}])[0].get("studentCountPerTeacher", "없음"),
                            }
                        except:
                            continue

                        # 최종 데이터 리스트에 추가
                        final_data.append(data_entry)
                        with open('apartment_info_final.jsonl', 'a', encoding='utf-8') as f:
                            for entry in final_data:
                                json.dump(entry, f, ensure_ascii=False)
                                f.write('\n')
                        print('저장 완료')
                        final_data = []
```
